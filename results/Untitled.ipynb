```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

ttt_sums = []
with open("ttt.result") as file:
    for line in file:
        nums = [float(s) for s in line.split()]
        ttt_sums.append(sum(nums))
kalaha_sums = []
with open("kalaha.result") as file:
    for line in file:
        nums = [float(s) for s in line.split()]
        kalaha_sums.append(sum(nums))

plt.figure(figsize=(15,3))
plt.plot(range(0, 1500, 50), ttt_sums, label='TicTacToe')
plt.plot(range(0, 1500, 50), kalaha_sums, label='Kalah')
plt.xlabel("Episodes")
plt.ylabel("Evaluation Score")
plt.title("Performance of AlphaZero")
plt.legend()
```
